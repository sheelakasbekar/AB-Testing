```python
# The counts of each parameter to compare
Reach_Count = data["Reach"].sum(), data1["Reach"].sum()
colors = ["green" , "red"]
plt.pie(Reach_Count, labels=['data', 'data1'],colors=colors,autopct ='1%.1f%%',startangle=90,)
plt.axis('Equal')
plt.title('Reach of Control Data VS Test Data')
plt.xlabel('Reach Count')
plt.show()
```
